```python
import pandas as pd

df = pd.read_csv('../Original Data/billboard.csv', encoding = 'latin')
df
```

```python
to_be_melted = list(df.columns[7:])
melted = df.melt(id_vars = ['artist.inverted', 'track', 'date.entered'], var_name = 'week', value_vars = to_be_melted, value_name = 'place')
music_detail = df.drop(columns=list(df.columns[5:]))
music_detail.sort_values(['artist.inverted'], inplace=True)
music_detail.reset_index(drop=True, inplace=True)
```

```python
melted.dropna(axis=0, inplace=True)
```

```python
melted.sort_values(['artist.inverted', 'date.entered', 'week'], inplace=True)
melted.reset_index(drop=True, inplace=True)
better_time = list(pd.to_datetime(melted['date.entered']))
melted['date.entered'] = better_time
melted['week'] = melted['week'].str.lstrip('x')
melted['week'] = melted['week'].str.rstrip('.week')
melted['week'] = melted['week'].str.rstrip('st|nd|rd|th')

```

```python
for i in range(len(melted)):
    melted.loc[i, 'date.entered'] = melted.loc[i, 'date.entered'] + pd.DateOffset(days = 7*(int(melted.loc[i, 'week']) - 1))
    melted.loc[i, 'week'] = melted.loc[i, 'date.entered'] + pd.DateOffset(days = 7)
melted
```

```python
melted['week'] = melted['week'].astype('datetime64[ns]')
melted = melted.rename(columns={'artist.inverted' : 'artist', 'date.entered' : 'week start', 'week' : 'week end'})
```

```python

melted.to_csv(r'..\Analysis Data\music_top_list.csv')
```
