```python
import pandas as pd

df = pd.read_csv('../Original Data/billboard.csv')
df
```

```python
to_be_melted = list(df.columns[7:])
melted = df.melt(id_vars = ['artist.inverted', 'track', 'date.entered'], var_name = 'week', value_vars = to_be_melted, value_name = 'place')
melted
```

```python
melted.dropna(axis=0, inplace=True)
melted
```

```python
melted.sort_values(['artist.inverted', 'date.entered', 'week'], inplace=True)
melted.reset_index(drop=True, inplace=True)
better_time = list(pd.to_datetime(melted['date.entered']))
melted['date.entered'] = better_time
melted['week'] = melted['week'].str.lstrip('x')
melted['week'] = melted['week'].str.rstrip('.week')
melted['week'] = melted['week'].str.rstrip('st|nd|rd|th')
melted
```
